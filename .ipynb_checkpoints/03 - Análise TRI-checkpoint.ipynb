## 2 -  Análise TRI via pacote ltm 
A análise via Teoria da Reposta ao Item é feita em duas etapas:

 * Calibração de Parâmetros de Itens
 * Estimação de Proficiência de Respondentes 

### 2.1 Calibraçção de parâmetros de itens

```r
# No pacote a função tmp estima parâmetros de itens usando o método da Máxima Verossimilhança
ltm_enem <- tpm(dplyr::select(amostra_correcao_ordenada,-NU_INSCRICAO))
```

```r
ltm_enem
```

```r
ltm_enem_pars <- coef(ltm_enem, simplify = TRUE)
head(ltm_enem_pars)
```

```r
# As Curvas Características dos 6 primeiros Itens  
options(repr.plot.width = 9, repr.plot.height = 6, repr.plot.res = 300)
plot(ltm_enem,type="ICC",items=1:6)
```

```r
# As Curvas Informação dos 6 primeiros Itens  
options(repr.plot.width = 9, repr.plot.height = 6, repr.plot.res = 300)
plot(ltm_enem,type="IIC",items=1:5)
```

```r
# A Curva Informação Total do Teste  
options(repr.plot.width = 9, repr.plot.height = 6, repr.plot.res = 300)
plot(ltm_enem,type="IIC",items=0)
```

### 2.2 Estimação de proficiência de respondentes

```r
# No pacote ltm podem ser utilizados os seguintes métodos de estimação de proficiências
# EB: Empirical Bayes
# MI: Multiple Imputation
# EAP: Expected a Posteriori

ltm_enem_profic <- factor.scores.tpm(ltm_enem, method = "EAP")
```

```r
head(ltm_enem_profic$score.dat)
```

```r
options(repr.plot.width = 9, repr.plot.height = 6, repr.plot.res = 300)
gghistogram(ltm_enem_profic$score.dat$z1)
```

## 3 -  Análise TRI via pacote mirt 
A análise via Teoria da Reposta ao Item é feita em duas etapas:

 * Calibração de Parâmetros de Itens
 * Estimação de Proficiência de Respondentes 

### 3.1 Calibração de parâmetros de itens

```r
# No pacote mirt podem ser utilizados os seguintes métodos de estimação de parâmetros de itens
# EM: EM algorithm with fixed quadrature (padrão)
# QMCEM: quasi-Monte Carlo EM estimation
# MCEM: Monte Carlo EM estimation
# MHRM: MH-RM algorithm
# SEM: Stochastic EM algorithm
# BL: Bock and Lieberman approach
mirt_enem <- mirt(
    data = dplyr::select(amostra_correcao_ordenada,-NU_INSCRICAO), 
    model = 1, 
    itemtype= '3PL',
    method = 'EM', 
    SE=T, 
    technical = list(NCYCLES = 1000))
```

```r
mirt_enem_pars = coef(mirt_enem, IRTpars=TRUE, simplify=TRUE)$items
```

```r
head(mirt_enem_pars)
```

```r
# Plot da Curva Característica e Informação do Item
itemplot(mirt_enem, item = "I_005961", type = 'infotrace') 
```

```r
# Plot da Curva de Informação Total do Teste
plot(mirt_enem, type="info")
```

### 3.2 Estimação de proficiência de respondentes

```r
# No pacote `mirt` podem ser utilizados os seguintes métodos de estimação de proficiências
# EAP: Expected a Posteriori (Padrão)
# MAP: Maximum a Posteriori
# ML: Maximum Likelihood
# WLE: Weighted Likelihood Estimation
# EAPsum: Expected a Posteriori for each sum score
# plausible: Single plausible value imputation for each case
mirt_enem_profic <-  fscores(mirt_enem, method = "EAP",response.pattern = dplyr::select(amostra_correcao_ordenada,-NU_INSCRICAO)) 
```

```r
mirt_enem_profic <- data.frame(mirt_enem_profic)
head(mirt_enem_profic)
```

```r
options(repr.plot.width = 9, repr.plot.height = 6, repr.plot.res = 300)
gghistogram(mirt_enem_profic$F1)
```

## 4 - Seleção de parâmetros de itens  

Antes da etapa de estimação de habilidades, é importante definir quais são os itens mais adequados para a avaliação.
A etapa de calibração de parâmetros de itens deve ser realizada cuidadosamente, respeitando os resultados encontrados na análise TCT.


```r
# Funções para formatar a exibição dos parâmetros segundo seus respectivos critérios
check_par_a <- function(a){
  a <- format(round(a,4), nsmall = 4)
  cell_spec(a,
            bold=ifelse(as.numeric(a) < 0.2, T,F),
            color=ifelse(as.numeric(a) < 0.2, "red","black"))
  
}
check_par_b <- function(b){
  b <- format(round(b,4), nsmall = 4)
  cell_spec(b,
            bold=ifelse(as.numeric(b) < -7 | as.numeric(b) > 7, T,F),
            color=ifelse(as.numeric(b) < -7 | as.numeric(b) > 7, "red","black"))
}
check_par_c <- function(c){
  c <- format(round(c,4), nsmall = 4)
  cell_spec(c,
            bold=ifelse(as.numeric(c) > 0.45, T,F),
            color=ifelse(as.numeric(c) > 0.45, "red","black"))
}
```

 #### 4.1 - Rodada 1

```r
itens_rodada_1 <- colnames(dplyr::select(amostra_correcao_ordenada,-NU_INSCRICAO))
```

```r
mod_rodada_1 <- mirt(dplyr::select(amostra_correcao_ordenada,all_of(itens_rodada_1)), 1, itemtype= '3PL', SE=T, technical = list(NCYCLES = 1000))
```

```r
pars_mod_rodada_1 = coef(mod_rodada_1, IRTpars=TRUE, simplify=TRUE)$items[,1:3]

pars_mod_rodada_1 |> 
  data.frame() |> 
  rownames_to_column("Item") |>
  rename(Dificuldade = b,Discriminacao = a, Chute = g) |> 
  mutate(Dificuldade    = check_par_b(Dificuldade)) |> 
  mutate(Discriminacao  = check_par_a(Discriminacao)) |> 
  mutate(Chute          = check_par_c(Chute)) |> 
  kbl(linesep="", escape = F,row.names=F,align="c",digits = 4,longtable = T) |> 
  as.character() |>
  display_html()
```

#### 4.2 - Rodada 2
 Na Rodada 1, alguns parâmetros apresentaram valores inadequados, precisamos removê-los.

```r
# Selecionando itens da Rodada 2
itens_rodada_2 <- amostra_correcao_ordenada |> 
  dplyr::select(-NU_INSCRICAO, -I_086499,-I_086840,-I_089637,-I_117820,-I_117886) |> 
  colnames()
```

```r
mod_rodada_2 <- mirt(dplyr::select(amostra_correcao_ordenada,all_of(itens_rodada_2)), 1, itemtype= '3PL', SE=T, technical = list(NCYCLES = 1000))
```

```r
pars_mod_rodada_2 = coef(mod_rodada_2, IRTpars=TRUE, simplify=TRUE)$items[,1:3]

pars_mod_rodada_2 |> 
  data.frame() |> 
  rownames_to_column("Item") |>
  rename(Dificuldade = b,Discriminacao = a, Chute = g) |> 
  mutate(Dificuldade    = check_par_b(Dificuldade)) |> 
  mutate(Discriminacao  = check_par_a(Discriminacao)) |> 
  mutate(Chute          = check_par_c(Chute)) |> 
  kbl(linesep="", escape = F,row.names=F,align="c",digits = 4,longtable = T) |> 
  as.character() |>
  display_html()
```


#### 4.3 - Estimação de Proficiências
Por fim, podemos realizar a estimação de proficiências.

```r
proficiencia <-  fscores(
    mod_rodada_2, 
    method = "EAP",
    response.pattern = dplyr::select(amostra_correcao_ordenada,all_of(itens_rodada_2))
) 
```

```r
options(repr.plot.width = 9, repr.plot.height = 6, repr.plot.res = 300)
gghistogram(proficiencia[,1],xlim=c(-3,3),binwidth = 0.25,xlab = "Proficiência")
```
